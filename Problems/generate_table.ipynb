```python

import matplotlib.pyplot as plt
from matplotlib import rcParams
import textwrap

# ----------------------------#
# Configuration
# ----------------------------#
rcParams["font.family"] = "sans-serif"
rcParams["font.size"] = 9

wrap_width = 40  # characters before wrapping long text
table_font_size = 9

# ----------------------------#
# Data
# ----------------------------#
headers1 = [
    "HEA ID #",
    "Composition (atomic %)",
    "Fracture-Toughness Index (MPa·√m)",
]
rows1 = [
    [
        "15",
        "chromium 33.33, cobalt 33.33, nickel 33.33",
        "265.20",
    ],
    [
        "98",
        "aluminum 20, chromium 20, cobalt 20, iron 20, nickel 20",
        "186.47",
    ],
    [
        "1",
        (
            "aluminum 10.53, boron 5.26, chromium 10.53, cobalt 10.53, "
            "copper 10.53, iron 10.53, molybdenum 10.53, nickel 10.53, "
            "silicon 10.53, titanium 10.53"
        ),
        "50.90",
    ],
    [
        "53",
        "molybdenum 25, niobium 25, tantalum 25, tungsten 25",
        "2.90",
    ],
    [
        "104",
        "cobalt 20, chromium 20, iron 20, manganese 20, titanium 20",
        "2.44",
    ],
]

headers2 = [
    "Air Temperature",
    "Wind-chill temperature\nat wind speed 15 mph",
    "Wind-chill temperature\nat wind speed 45 mph",
]
rows2 = [
    ["26° F", "14° F", "7° F"],
    ["30° F", "19° F", "12° F"],
    ["34° F", "24° F", "18° F"],
]


# ----------------------------#
# Helper to wrap long strings
# ----------------------------#
def wrap_row(row, indices_to_wrap, width=wrap_width):
    """Return a new row with selected indices wrapped to the specified width."""
    new_row = row.copy()
    for idx in indices_to_wrap:
        new_row[idx] = textwrap.fill(new_row[idx], width=width)
    return new_row


# ----------------------------#
# Figure 1 – Mechanical-Property Data
# ----------------------------#
fig1, ax1 = plt.subplots(figsize=(8, 10))
ax1.axis("off")

celltext1 = [
    wrap_row(r, indices_to_wrap=[1], width=wrap_width) for r in rows1
]

table1 = ax1.table(
    cellText=celltext1,
    colLabels=headers1,
    cellLoc="left",
    loc="center",
)
table1.auto_set_font_size(False)
table1.set_fontsize(table_font_size)
table1.auto_set_column_width(col=list(range(len(headers1))))

# Title
ax1.set_title(
    "Table 1. Mechanical-Property Data",
    fontweight="bold",
    pad=20,
    fontsize=11,
)

plt.tight_layout()

# ----------------------------#
# Figure 2 – Wind-Chill Reference
# ----------------------------#
fig2, ax2 = plt.subplots(figsize=(6, 4))
ax2.axis("off")

table2 = ax2.table(
    cellText=rows2,
    colLabels=headers2,
    cellLoc="center",
    loc="center",
)
table2.auto_set_font_size(False)
table2.set_fontsize(table_font_size)
table2.auto_set_column_width(col=list(range(len(headers2))))

# Title
ax2.set_title(
    "Table 2. Wind-Chill Temperatures",
    fontweight="bold",
    pad=15,
    fontsize=11,
)

plt.tight_layout()

# ----------------------------#
# Display or save
# ----------------------------#
plt.show()

# Uncomment below lines if you prefer saving to PNG
# fig1.savefig("table1_mechanical_properties.png", dpi=300, bbox_inches="tight")
# fig2.savefig("table2_wind_chill.png", dpi=300, bbox_inches="tight")

```
